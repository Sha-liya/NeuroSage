```python
y_reg = df["MMSE"]
```

```python
y_clf = df["Diagnosis"]
```

```python
X = df.drop(columns=["MMSE", "Diagnosis"])
```

```python
X = df.drop(columns=[
    "PatientID",
    "DoctorInCharge",
    "MMSE",
    "Diagnosis"
])

X.head()
```

```python
print(X.select_dtypes(include=["object"]).columns)
```

```python
print(X.select_dtypes(include=["category"]).columns)
```

```python
from sklearn.model_selection import train_test_split

# Regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X,
    y_reg,
    test_size=0.2,
    random_state=42
)

# Classification
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X,
    y_clf,
    test_size=0.2,
    random_state=42
)
```

```python
print(X_train_reg.shape)
print(X_test_reg.shape)

print(y_train_reg.shape)
print(y_test_reg.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_reg_scaled = scaler.fit_transform(X_train_reg)
X_test_reg_scaled = scaler.transform(X_test_reg)

X_train_clf_scaled = scaler.fit_transform(X_train_clf)
X_test_clf_scaled = scaler.transform(X_test_clf)
```

```python
print(X.select_dtypes(include=["object"]).columns)
```

```python
print(X_train_reg.shape)
print(X_test_reg.shape)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()

linear_model.fit(X_train_reg_scaled, y_train_reg)
```

```python
y_pred_reg = linear_model.predict(X_test_reg_scaled)

print(y_pred_reg[:10])
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test_reg, y_pred_reg)
r2 = r2_score(y_test_reg, y_pred_reg)

print("Mean Squared Error (MSE):", mse)
print("R² Score:", r2)
```

```python
coefficients = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": linear_model.coef_
})

coefficients
```

```python
top3 = coefficients.reindex(
    coefficients["Coefficient"].abs().sort_values(ascending=False).index
).head(3)

print(top3)
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=1.0)

ridge_model.fit(X_train_reg_scaled, y_train_reg)

y_pred_ridge = ridge_model.predict(X_test_reg_scaled)

ridge_mse = mean_squared_error(y_test_reg, y_pred_ridge)

ridge_r2 = r2_score(y_test_reg, y_pred_ridge)

print("Ridge Regression")
print("MSE:", ridge_mse)
print("R²:", ridge_r2)
```

```python
comparison = pd.DataFrame({
    "Model": ["Linear Regression", "Ridge Regression"],
    "MSE": [mse, ridge_mse],
    "R²": [r2, ridge_r2]
})

comparison
```

```python
print("Training Class Distribution")
print(y_train_clf.value_counts())

print("\nPercentage")
print(y_train_clf.value_counts(normalize=True) * 100)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)

log_model.fit(X_train_clf_scaled, y_train_clf)
```

```python
y_pred_clf = log_model.predict(X_test_clf_scaled)

y_pred_prob = log_model.predict_proba(X_test_clf_scaled)[:, 1]
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test_clf, y_pred_clf)

print(cm)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test_clf, y_pred_clf))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

accuracy = accuracy_score(y_test_clf, y_pred_clf)
precision = precision_score(y_test_clf, y_pred_clf)
recall = recall_score(y_test_clf, y_pred_clf)
f1 = f1_score(y_test_clf, y_pred_clf)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, thresholds = roc_curve(y_test_clf, y_pred_prob)

auc = roc_auc_score(y_test_clf, y_pred_prob)

print("AUC:", auc)

plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
plt.plot([0,1], [0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
thresholds = [0.30, 0.40, 0.50, 0.60, 0.70]

results = []

for t in thresholds:
    preds = (y_pred_prob >= t).astype(int)

    p = precision_score(y_test_clf, preds)
    r = recall_score(y_test_clf, preds)
    f = f1_score(y_test_clf, preds)

    results.append([t, p, r, f])

threshold_table = pd.DataFrame(
    results,
    columns=["Threshold", "Precision", "Recall", "F1"]
)

threshold_table
```
