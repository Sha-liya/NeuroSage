```python
skewness = df.select_dtypes(include=["int64", "float64"]).skew()

skewness.sort_values(key=abs, ascending=False)
```

```python
highest_skew = skewness.abs().idxmax()

print("Most skewed column:", highest_skew)
print("Skewness value:", skewness[highest_skew])
```

```python
columns = ["BMI", "MMSE"]

for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print("=" * 40)
    print(f"Column: {col}")
    print(f"Q1: {Q1:.2f}")
    print(f"Q3: {Q3:.2f}")
    print(f"IQR: {IQR:.2f}")
    print(f"Lower Bound: {lower:.2f}")
    print(f"Upper Bound: {upper:.2f}")
    print(f"Number of Outliers: {len(outliers)}")
```

```python
# Top two columns with highest absolute skewness
top2 = skewness.abs().sort_values(ascending=False).head(2).index

for col in top2:
    print("=" * 40)
    print(col)
    print("Mean   :", df[col].mean())
    print("Median :", df[col].median())
```

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

pearson = numeric_df.corr(method="pearson")
spearman = numeric_df.corr(method="spearman")

print("Pearson Correlation Matrix")
display(pearson)

print("Spearman Correlation Matrix")
display(spearman)
```

```python
difference = (spearman - pearson).abs()

difference
```

```python
pairs = (
    difference.unstack()
    .sort_values(ascending=False)
)

# Remove self-comparisons
pairs = pairs[pairs != 0]

# Remove duplicate pairs (A,B) and (B,A)
seen = set()
top_pairs = []

for (a, b), value in pairs.items():
    if (b, a) not in seen:
        seen.add((a, b))
        top_pairs.append(((a, b), value))
    if len(top_pairs) == 3:
        break

print("Top 3 Pearson vs Spearman Differences:")
for pair, value in top_pairs:
    print(pair, "Difference:", round(value, 4))
```

```python
grouped = df.groupby("Gender")["Age"].agg(["mean", "std", "count"])

grouped
```
